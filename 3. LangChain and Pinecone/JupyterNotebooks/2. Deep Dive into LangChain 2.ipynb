# Simple Chains

A sequence of step or action. Chains allow us to combine components to solve a task.

```python
from langchain.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain.chains import LLMChain

template = '''You are an experient musician.
Write a few sentences about the playing {composer} with {instrument}'''
prompt_template = PromptTemplate.from_template(template=template)

chain = LLMChain(llm=llm,
                 prompt=prompt_template,
                 verbose=True)
output = chain.invoke({'composer':'vivaldi', 'instrument' : 'electric guitar'})
print(output)
```

```python
# We can create input via argument instead of dictionary

template = "What is the capital of {country}. What are the top 3 places to visit and 3 top foods to try in that capital? Please give bulletpoints."
prompt_template = PromptTemplate.from_template(template=template)

chain = LLMChain(llm=llm,
                 prompt=prompt_template,
                 verbose=True)
country = input("Please enter the name of a Country")
output = chain.invoke(country)
print(output)
```

# Sequential Chains
You can make one or more calls to LLM and chain them to use output of one to the input of other.

```python
from langchain import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain.chains import LLMChain, SimpleSequentialChain
```

```python
llm1 = ChatOpenAI(model_name='gpt-3.5-turbo',temperature=0.5)
prompt_template1 = PromptTemplate.from_template(template="You are an expert Python programmer. Write a function that implements {concept}")
chain1 = LLMChain(llm=llm1,prompt=prompt_template1)

llm2 = ChatOpenAI(model_name='gpt-4-turbo-preview',temperature=1.2)
prompt_template2 = PromptTemplate.from_template(template="Given the Python function {function}, describe it in details.")
chain2 = LLMChain(llm=llm2,prompt=prompt_template2)

overall_chain = SimpleSequentialChain(chains=[chain1,chain2], verbose=True)
output= overall_chain.invoke('dynamic programming')
```

# Reasoning and Acting : ReAct
ReAct is a new approac that combines reasoning (chain of thoughts) and acting capabilities of LLMs

With ReAct LLM s generate reasoning traces and task-specific actions in an interleaved manner. 


```python
!pip install langchainhub -q
```

```python
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv(),override=True)
```

```python
from langchain.prompts import PromptTemplate
from langchain import hub
from langchain.agents import Tool, AgentExecutor, initialize_agent, create_react_agent
from langchain.tools import DuckDuckGoSearchRun, WikipediaQueryRun
from langchain.utilities import WikipediaAPIWrapper
from langchain_experimental.tools.python.tool import PythonREPLTool
from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI()
template = '''
Answer the folowing questions the best way possible:
Question: {q}
'''
prompt_template = PromptTemplate.from_template(template)
prompt = hub.pull('hwchase17/react')
print(type(prompt))
print(prompt.input_variables)
print(prompt.template)
```

```python
# 1 Pythion REPL to run Python code
python_repl = PythonREPLTool()
python_repl_tool = Tool(
    name="Python Repl",
    func=python_repl.run,
    description="Useful when you need to use Python code to answer a question. You should input Python code")

# 2 Wikipedia tool for searching 
api_wrapper = WikipediaAPIWrapper()
wikipedia = WikipediaQueryRun(api_wrapper=api_wrapper)
wikipedia_tool= Tool(
    name="Wikipedia",
    func=wikipedia.run,
    description = "Useful when you want to check some topic, person, place on Wikipedia")
# 3 Duck duck go for general searches
search= DuckDuckGoSearchRun()
duckduckgo_tool = Tool(
    name="Duck Duck Go Search",
    func=search.run,
    description="Useful when you need to perform an internet search to find the information that another tool does not have as response for")

tools=[wikipedia_tool, duckduckgo_tool, python_repl_tool]
agent = create_react_agent(llm,tools,prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools,verbose=True, handle_parsing_errors=True, max_iterations=10)
```

```python
question = "What are the first 20 numbers in the fibonacci sequence"
output = agent_executor.invoke({"input": prompt_template.format(q=question)})
```

```python
question = "Who is the former president of Brasil"
output = agent_executor.invoke({"input": prompt_template.format(q=question)})
```

```python
question = "Tell me about Napoleon early life"
output = agent_executor.invoke({"input": prompt_template.format(q=question)})
```

# Embeddings
Embeddings are the core of LLM applications
Text embeddings are numeric representations of text and are used in NLP and ML tasks
The distance between two embeddings / vectors measures their relateness
Similar embeddings or vectors represent similar topics
Vector databases are designed to store vector representation of data efficiently.

# Vector Database
Efficient data processing is a challenge
May of the latest AI application rely on vector embeddings - chatbots, qa systems and machine translation
CSV are not efficient enough for big data : specialized tools are developed, such as, Pinecone, Milvus, Qdrant, chroma.

# Vector Index and Namespaces 

Pinecone allows you to partition vectors in an index into nemespace.
Queries and other operations are scoped to a specific namespace, allowing for different request to search different subsets of your index.


## Namespaces
Every index consists of one or more namespaces.
Each vector exists exactly in one namespace
Namespaces are uniquely identified by a namespace name
The default namespace is represented by the empty string and is used if no specific namespace is specified


```python
index.describe_index_stats()
index = pc.Index("langchain")

import random

vectors = [[random.random() for _ in range (1536)] for v in range(5)]
ids = list('abcde')
index.upsert(vectors=zip(ids,vectors)) #default namespace
```

```python
vectors = [[random.random() for _ in range (1536)] for v in range(3)]
ids = list('xyz')
index.upsert(vectors=zip(ids,vectors),namespace="first_namespace")
```

```python
vectors = [[random.random() for _ in range (1536)] for v in range(2)]
ids = list('qp')
index.upsert(vectors=zip(ids,vectors),namespace="second_namespace")
```

# Split and Index Text

There are several service loaders to add data to app: note https://python.langchain.com/v0.1/docs/modules/data_connection/document_loaders/

```python
import os
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv(), override=True)

print("Current pwd: " + os.getcwd())
print(os.path.exists("./files/churchill_speech.txt"))

from langchain.text_splitter import RecursiveCharacterTextSplitter # splits on 3 types of chars: \\n \n and whitespace
f = open("./files/churchill_speech.txt")
churchill_speech = f.read()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20,length_function=len)
chunks = text_splitter.create_documents([churchill_speech])
print(chunks[0])
print(chunks[0].page_content)
print(f"Now we have {len(chunks)} chunks")
```

# Embedding Cost

```python
def print_embedding_cost(texts):
    import tiktoken
    enc = tiktoken.encoding_for_model("text-embedding-ada-002")
    total_tokens = sum([len(enc.encode(page.page_content)) for page in texts])
    print(f"Total tokens: {total_tokens}")
    print(f"Embedding Cost in USD: {total_tokens / 1000 * 0.0004:.6f} ")

print_embedding_cost(chunks)
```

```python
# Here we are going to use Open AI to embed the text information into vectors

# !pip install -U langchain-openai #latest version
from langchain.embeddings import OpenAIEmbeddings
embedding= OpenAIEmbeddings()

# how to embed a chunk
vector=embedding.embed_query(chunks[0].page_content)
print(vector)


```
